```python
#importo librerias
import pandas as pd

#Leo el dataset
df = pd.read_excel('Asistencia a servicios.xlsx')

# Veo las premeras filas del dataset
df.head()
```

```python
cantidad_columnas = len(df.columns)
columnas_categoricas = df.select_dtypes(include=['object', 'category']).shape[1]
columnas_numericas = df.select_dtypes(exclude=['object', 'category']).shape[1]
cantidad_filas = df.shape[0]
porcentaje_nulos_total = (df.isnull().sum().sum() / df.size) * 100
columnas_con_nulos = df.isnull().sum()[df.isnull().sum() > 0].shape[0]
```

## Hasta este punto, visualizo un dataset sin mucho faltante de datos que perjudique o desvie el analisis

```python
#una vez identificaddos los valores nulos, procedo a tratarlos.
#Decido rellenar dichos datos con la moda, ya que centro de salud y nombre de doctor no me afectan en la busqueda que estoy realizando


for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].fillna(df[column].mode()[0])

# Veo el número de valores nulos por columna después de rellenar
df.isnull().sum()

# Veo las primeras filas del dataframe limpio
df.head()
```

#Realizo un análisis univariado

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizo la distribución del precio de las consultas
plt.figure(figsize=(10,6))
sns.histplot(df['Precio de la consulta'], kde=True)
plt.title('Distribución del Precio de la Consulta')
plt.show()

# Análisis univariado de las variables categóricas
categorical_columns = ['Genero del paciente', 'Tipo de servicio', 'Proveedor del seguro']

# Visualizo las distribuciones de las variables categóricas
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=df[column], palette='Set2')
    plt.title(f'Distribución de la variable {column}')
    plt.xticks(rotation=45)
    plt.show()

# Análisis univariado de las variables numéricas
numerical_columns = ['Precio de la consulta', 'Edad del paciente', 'Minutos de la consulta']

# Estadísticas descriptivas para las variables numéricas
print("\nEstadísticas descriptivas para las variables numéricas:")
print(df[numerical_columns].describe())

# Visualización de la distribución de las variables numéricas
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], kde=True, color='skyblue')
    plt.title(f'Distribución de {column}')
    plt.show()
```

#Realizo un análisis bivariado

```python
# Análisis bivariado entre variables categóricas y numéricas
for column in categorical_columns:
    for num_column in numerical_columns:
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=df, x=column, y=num_column, palette='Set2')
        plt.title(f'Análisis Bivariado: {num_column} vs {column}')
        plt.xticks(rotation=45)
        plt.show()


# Correlación entre variables numéricas
correlation_matrix = df[numerical_columns].corr()

# Visualización de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.show()

# Visualizo la distribución de la Edad del Paciente vs Precio de la Consulta con un KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Edad del paciente', y='Precio de la consulta', cmap='Blues', shade=True, shade_lowest=False)
plt.title('Densidad de Edad del Paciente vs Precio de la Consulta')
plt.show()



# Visualizo la relación entre el tipo de servicio y el precio
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Tipo de servicio', y='Precio de la consulta')
plt.title('Precio de la Consulta por Tipo de Servicio')
plt.xticks(rotation=45)
plt.show()
```

#Veo por lo pronto que la cantidad de atenciones se distribuye de manera normal por precio. Por otro lado , observo que los precios mas altos en cantidad y en valor se presentan en un leve aumento entre lo 20 y 30 años, mientras que los precios de servicios mas bajos se concentran aprox en los 80 años, lo cual refuta mi hipotesis planteada en principio. Por ultimo, no hay variacion de precio por tipo de servicio, lo cual apunta a una posible revision por parte de los establecimientos y los proveedores de seguros para redefinir los precios

#Analizo y grafico presencia de outliers

```python

numerical_columns = ['Precio de la consulta', 'Edad del paciente', 'Minutos de la consulta']

# Calcular Q1, Q3 y el IQR para cada una de las columnas numéricas
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Identificar outliers
    outliers = df[(df[column] < limite_inferior) | (df[column] > limite_superior)]

    print(f"\nAnálisis de la variable '{column}':")
    print(f"  - Primer Cuartil (Q1): {Q1}")
    print(f"  - Tercer Cuartil (Q3): {Q3}")
    print(f"  - Rango Intercuartílico (IQR): {IQR}")
    print(f"  - Límite Inferior: {limite_inferior}")
    print(f"  - Límite Superior: {limite_superior}")
    print(f"  - Número de Outliers: {outliers.shape[0]}")

```

```python
# Visualizo outliers en el precio de la consulta usando boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Precio de la consulta')
plt.title('Outliers en el Precio de la Consulta')
plt.show()
```

```python
# Visualizar outliers en la edad del paciente usando boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Edad del paciente')
plt.title('Outliers en la Edad del Paciente')
plt.show()
```

```python
# Visualizo outliers en los minutos de la consulta usando boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Minutos de la consulta')
plt.title('Outliers en los Minutos de la Consulta')
plt.show()
```

#se verifica que no hay presencia de outliers, por lo que la muestra arroja resultados sin desvios y no es necesario por lo pronto seguir limpiando la muestra para ajustar el resultado.
